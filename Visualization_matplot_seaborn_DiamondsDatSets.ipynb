```python
#look into columns like carat,depth,price using matplotlib
fig=plt.figure(figsize=(20,20))

#1)carat weight distribution
plt.subplot(221)
plt.hist(diamonds['carat'],bins=20,color='b')
plt.xlabel('Carat Weight')
plt.ylabel('Frequency')
plt.title('Carat weight Distribution Plot')
```

```python
#2)Price Distribution Plot
fig=plt.figure(figsize=(20,20))
plt.subplot(222)
plt.hist(diamonds['price'],bins=20,color='y')
plt.xlabel('Price of Diamond')
plt.ylabel('Frequency')
plt.title('Price Distribution Plot')
```

```python
diamonds['Volume']=diamonds['x']*diamonds['y']*diamonds['z']
diamonds
```

```python
plt.figure(figsize=[12,12])
```

```python
#2)volume Distribution Plot
fig=plt.figure(figsize=(40,40))
plt.subplot(223)
plt.hist(diamonds['Volume'],bins=20,color='r')
plt.xlabel('Volume of Diamond')
plt.ylabel('Frequency')
plt.title('Volume Distribution Plot')
```

```python
#4)depth Distribution Plot
fig=plt.figure(figsize=(20,20))
plt.subplot(224)
plt.hist(diamonds['depth'],bins=20,color='y')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.title('Depth Distribution Plot')
```
